```python
import pandas as pd
```

```python
df = pd.read_csv('/content/tcc_ceds_music.csv')
```

```python
df.head()
```

```python
unique_genres = df['genre'].unique()
print(unique_genres)
```

```python
country_rows = df[df['genre'] == 'country']
```

```python
lyrics = country_rows["lyrics"]
print(lyrics)
```

```python
lyrics_column = country_rows['lyrics']

# Count the words in each row of the 'lyrics' column and store the counts in a new column 'count'
country_rows['count'] = lyrics_column.apply(lambda x: len(str(x).split()))

# Now, keep only the 'lyrics' and 'count' columns
country_lyrics_with_count = country_rows[['lyrics', 'count']]

# Print the resulting DataFrame
print(country_lyrics_with_count)
```

```python
lyrics = country_rows["lyrics"]
lyrics.head()
```

```python
file_path = 'lyrics.csv'

# Save the DataFrame to a CSV file
lyrics.to_csv(file_path, index=False)

print(f"DataFrame saved to {file_path}")
```
